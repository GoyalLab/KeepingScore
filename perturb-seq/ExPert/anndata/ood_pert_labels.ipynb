# OOD perturbation label names

Map the integer labels in `splits/y_ood.npy` to the actual perturbation names in `splits/perturbation_names.npy`. The `ood_label_0_92` column is a compact index over the 93 OOD perturbation types; `original_label` is the integer label stored in `y_ood.npy`.

```python
from pathlib import Path
import csv
import numpy as np

split_dir = Path('/projects/b1042/GoyalLab/jaekj/KeepingScore/perturb-seq/ExPert/anndata/splits')
y_ood = np.load(split_dir / 'y_ood.npy', allow_pickle=True).astype(int).ravel()
perturbation_names = np.load(split_dir / 'perturbation_names.npy', allow_pickle=True)

ood_original_labels = sorted(np.unique(y_ood).tolist())
ood_label_to_name = [
    {
        'ood_label_0_92': ood_label,
        'original_label': original_label,
        'perturbation_name': str(perturbation_names[original_label]),
    }
    for ood_label, original_label in enumerate(ood_original_labels)
]

print(f'Number of OOD perturbation types: {len(ood_label_to_name)}')
print('ood_label_0_92\toriginal_label\tperturbation_name')
for row in ood_label_to_name:
    print(f"{row['ood_label_0_92']}\t{row['original_label']}\t{row['perturbation_name']}")

out_csv = split_dir / 'ood_label_to_perturbation_name.csv'
with out_csv.open('w', newline='') as handle:
    writer = csv.DictWriter(handle, fieldnames=['ood_label_0_92', 'original_label', 'perturbation_name'])
    writer.writeheader()
    writer.writerows(ood_label_to_name)

out_csv
```
